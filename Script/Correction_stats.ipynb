```python
import os
import json



def calcola_percentuale_corretto(cartella):
    totale = 0
    corretti = 0
    for nome_file in os.listdir(cartella):
        if nome_file.endswith('.json'):
            percorso_file = os.path.join(cartella, nome_file)
            with open(percorso_file, 'r', encoding='utf-8') as f:
                try:
                    dati = json.load(f)
                except json.JSONDecodeError:
                    continue
                for dato in dati:
                        totale += 1
                        if dato['correct'] is True:
                            corretti += 1
            percentuale = (corretti / totale * 100) if totale > 0 else 0
            if "note" in nome_file:
                nomeapp = "note"
            elif "tfa" in nome_file:
                nomeapp = "tfa"
            else:
                nomeapp = "pass"

            if "Humanoid" in cartella:
                nomeapp += "-humanoid"
            if "Replay" in cartella:
                nomeapp += "-replay"
            if "LLAMA" in cartella:
                nomeapp += "-llama"
            if "Scelta" in cartella:
                nomeapp += "-scelta"
            if "prelogin" in nome_file:
                nomeapp += "-prelogin"
            elif "postlogin" in nome_file:
                nomeapp += "-postlogin"
            print(f"Analisi file: {nome_file} in {cartella}")
            print(f"Percentuale correttezza per {nomeapp} in {cartella_input}: {percentuale:.2f}% ({corretti}/{totale})")

# Esempio di utilizzo:
cartelle = [r"01-Droidbot Replay Output",r"02-Droidbot + LLAMA",r"03-Droidbot + LLAMA + Scelta", r"05-Humanoid + Replay Output", r"06-Humanoid + LLAMA"]

for cartella_input in cartelle:
    print(f"Analisi file in {cartella_input}" )
    #calcola_percentuale_corretto(os.path.join(r"C:\Users\juve2\Desktop\Projects\Project_LLM_A3\Result",cartella_input))
    calcola_percentuale_corretto(os.path.join(r"D:\documenti\UNI\LLM\progetto\git\Project_LLM_A3\Result",cartella_input))
    print("\n" +"\n")
    
```

```python
### Temporaneo , aggiunto da Mike

import os
import json

def calcola_percentuale_corretto_singolo_file(cartella, nome_file):
    percorso_file = os.path.join(cartella, nome_file)
    totale = 0
    corretti = 0

    if os.path.isfile(percorso_file):
        with open(percorso_file, 'r', encoding='utf-8') as f:
            try:
                dati = json.load(f)
            except json.JSONDecodeError:
                print(f"Errore nel decodificare il file {percorso_file}")
                return
            
            for dato in dati:
                totale += 1
                if dato.get('correct') is True:
                    corretti += 1
        
        percentuale = (corretti / totale * 100) if totale > 0 else 0
        
        nomeapp = "note"
        if "Humanoid" in cartella:
            nomeapp += "-humanoid"
        
        print(f"Analisi file: {nome_file} in {cartella}")
        print(f"Percentuale correttezza per {nomeapp} in {cartella}: {percentuale:.2f}% ({corretti}/{totale})")
    else:
        print(f"File {nome_file} non trovato nella cartella {cartella}")

# Uso:
cartella_input = r"C:\Users\miche\Desktop\Project LLM\Project_LLM_A3\Result\06-Humanoid + LLAMA"
nome_file = "risultati_output-note.json"

calcola_percentuale_corretto_singolo_file(cartella_input, nome_file)
```
