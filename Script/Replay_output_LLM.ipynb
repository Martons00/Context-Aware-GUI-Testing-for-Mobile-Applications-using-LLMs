```
import os
import json
import torch
from transformers import pipeline, AutoModelForCausalLM, AutoTokenizer

from huggingface_hub import login


# Standard library
import os
os.environ['PYTORCH_CUDA_ALLOC_CONF'] = 'expandable_segments:True'
from huggingface_hub import login
from google.colab import userdata

HF_TOKEN = userdata.get('token_llama')
if HF_TOKEN:
    login(HF_TOKEN)
    print("Successfully logged in to Hugging Face!")
else:
    print("Token is not set. Please save the token first.")

model_id = "meta-llama/Llama-3.2-3B-Instruct"
model = AutoModelForCausalLM.from_pretrained(model_id, torch_dtype=torch.float16, device_map="auto")
tokenizer = AutoTokenizer.from_pretrained(model_id)

# Assuming model and tokenizer are already loaded
device = torch.device("cuda" if torch.cuda.is_available() else "cpu")

# Move model to the device (GPU if available)
model = model.to(device)



def create_prompt(event):
    view = event.get("view", {})
    context_parts = []

    for key in ["class", "resource_id", "content_description"]:
        value = view.get(key)
        if value:
            context_parts.append(f"{key}={value}")

    context = ", ".join(context_parts) if context_parts else "Campo testo generico"
    prompt = f"""{INITIAL}

    {FEW_SHOTS}

Context: {context}
Input:"""
    prompt += "Campo Generico" if not view.get("resource_id") else view.get("resource_id").split("/")[-1]

    return prompt

i=0

def generate_response(prompt):
    global i
    # Tokenize the input

    prompt += "output : "
    inputs = tokenizer(prompt, return_tensors="pt").to('cuda')
    output = model.generate(
        inputs['input_ids'],  # Tokenized input
        max_length=200,         # Limit response length to avoid extra text
        temperature=0.7,        # Lower temperature to reduce randomness
        do_sample=True,        # Disable sampling for deterministic output
        pad_token_id=tokenizer.eos_token_id  # Ensure the model doesn't go beyond the end token
    )
    response = tokenizer.decode(output[0], skip_special_tokens=True)
    i+=1
    print(f"Response {i}: \n Prompt: ")
    print(response.split("output :")[0].split("Input:")[-1].strip(), "\n Answer: ")
    print(response.split("output :")[-1].strip())
    return response.split("output : ")[-1].strip()


def modify_events_folder(events_dir1,events_dir2):
    for filename in os.listdir(events_dir1):
        if filename.endswith(".json"):
            file_path1 = os.path.join(events_dir1, filename)
            file_path2 = os.path.join(events_dir2,filename)
            with open(file_path1, 'r', encoding='utf-8') as f:
              try:
                  data = json.load(f)
              except json.JSONDecodeError:
                  print(f"Error decoding JSON in file: {file_path1}")
                  continue

              if isinstance(data, dict):
                event = data["event"]
                if data.get("event", {}).get("event_type") == "set_text":
                  prompt = create_prompt(event)
                  new_text = generate_response(prompt)
                  event["text"] = new_text
                  data["text"] = new_text
                with open(file_path2, 'w', encoding='utf-8') as f:
                  json.dump(data, f, indent=2)

def generate_set_text_script(events_dir, output_file):
    script = []
    for filename in os.listdir(events_dir):
        if filename.endswith(".json"):
            file_path = os.path.join(events_dir, filename)
            with open(file_path, 'r', encoding='utf-8') as f:
                data = json.load(f)
                if isinstance(data, dict) and data.get("event", {}).get("event_type") == "set_text":
                    event = data["event"]
                    prompt = create_prompt(event)
                    new_text = generate_response(prompt)
                    event["text"] = new_text
                    script.append(event)

    with open(output_file, 'w', encoding='utf-8') as f:
        json.dump(script, f, indent=2)


events_to_modify_folder = "/content/drive/MyDrive/LLM/events_preLogin"
events_output_folder = "/content/drive/MyDrive/LLM/droidbot_replay_tfa_pre"

if not os.path.exists(events_output_folder):
    os.makedirs(events_output_folder)

# Esecuzione (in Colab, `events` devessere gi montata da Drive)
#generate_set_text_script("/content/drive/MyDrive/LLM/events_postLogin", "/content/drive/MyDrive/LLM/set_text_script_postLogin.json")
modify_events_folder(events_to_modify_folder,events_output_folder)


print("pass done\n\n")


```
